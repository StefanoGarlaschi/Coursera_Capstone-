```python
import numpy as np
import pandas as pd
```

```python
#importing geospatial information for the neighborhoods
df_geo = pd.read_csv('Geospatial_Coordinates.csv')
df_geo.head(10)
```

```python
df_geo.shape
```

```python
#importing the dataframe for Toronto created in the first part
df = pd.read_csv('Toronto-table1.csv')
df.head(10)
```

```python
#appending latitude and longitude for the neighboorhood
latitude = []
longitude = []

for i in range(df.shape[0]):
    postal_code = df.PostalCode[i]
    
    temp = df_geo.loc[df_geo['Postal Code'] == postal_code].reset_index(drop=True)
    lat = temp['Latitude'][0]
    long = temp['Longitude'][0]
    latitude.append(lat)
    longitude.append(long)
    
df['Latitude'] = latitude
df['Longitude'] = longitude
df.head(10)
    
#print(len(latitude))
#print(len(longitude))
```

```python
df.to_csv('Toronto-table2.csv',index=False)
```
